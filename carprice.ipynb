```python
#significant variable after visual analysis

cars_lr=cars[['price', 'fueltype', 'aspiration','carbody', 'drivewheel','wheelbase',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 
                    'fueleconomy', 'carlength','carwidth', 'carsrange']]

cars_lr.head()
```

```python
sns.pairplot(cars_lr)
plt.show()
```

```python
#finding dummy 

def dummies (x,df):
    temp=pd.get_dummies(df[x],drop_first=True)
    df=pd.concat([df,temp], axis=1)
    df.drop([x], axis=1 ,inplace=True)
    return df


cars_lr=dummies('fueltype',cars_lr)
cars_lr=dummies('aspiration',cars_lr)
cars_lr=dummies('carbody',cars_lr)
cars_lr=dummies('drivewheel',cars_lr)
cars_lr=dummies('enginetype',cars_lr)
cars_lr=dummies('cylindernumber',cars_lr)
cars_lr=dummies('carsrange',cars_lr)

```

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(cars_lr, train_size=0.7, test_size=0.3 , random_state=100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
num_vars=['price','wheelbase','curbweight','enginesize','boreratio','horsepower','fueleconomy','carlength','carwidth']
df_train[num_vars]=scaler.fit_transform(df_train[num_vars])
```

```python
#dividing data into X and Y variable

y_train= df_train.pop('price')
x_train= df_train
```

```python
#model building
```

```python
# rfe ---> recursing feature elimination
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor as vv
```

```python
lm= LinearRegression()
lm.fit(x_train,y_train)
rfe=RFE(lm,10)
rfe=rfe.fit(x_train,y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
x_train.columns[rfe.support_]
```

```python
x_train_rfe=x_train[x_train.columns[rfe.support_]]
```

```python
x_train_rfe.head()
```

```python
def build_model(x,y):
    x=sm.add_constant(x)
    lm=sm.OLS(y,x).fit()   #ordinary least square
    print(lm.summary())
    return x

def checkVIF(x):
    vif=pd.DataFrame()
    vif['Features']=x.columns
    vif['VIF']=[vv(x.values,i) for i in range(x.shape[1])]
    vif['VIF']=round(vif['VIF'],2)
    vif=vif.sort_values(by = 'VIF' , ascending=False)
    return (vif)
    
```
